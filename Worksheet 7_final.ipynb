```python
# Ques f :Calculate what percentage of peoples were survived or not survived after their operation.
print("Percentage of survived :",len(haberman[haberman["Survival_status"]=="Yes"]) * 100 /len(haberman))
print("Percentage of Not Survived :",len(haberman[haberman["Survival_status"]=="No"]) * 100 /len(haberman))
```

```python
# Ques g :Draw a bar chart to show the survival or non survival distribution in the harberman cancer dataset.
sns.set(font_scale=1.4)
haberman['Survival_status'].value_counts().plot(kind='bar', figsize=(9, 6), rot=45)
plt.xlabel("Survival Status", labelpad=14)
plt.ylabel("Total Number", labelpad=14)
plt.title("Suvived vs Not Survived");
```

```python
# Ques h Draw a bar chart to show the survival or non survival distribution in the harberman cancer dataset. Give a different color for each distribution on the chart.
sns.set(font_scale=1.4)
haberman['Survival_status'].value_counts().plot(kind='bar', color=["g","r"],figsize=(9, 6), rot=90)
plt.xlabel("Survival Status", labelpad=14)
plt.ylabel("Total Number", labelpad=14)
plt.title("Suvived vs Not Survived");
```
